```python
y=dataframe.Salary
```

```python
f=['10percentage','12graduation', '12percentage','CollegeTier',]
 
```

```python
X= dataframe[f]
```

```python
train_X, val_X, train_y, val_y = train_test_split(X, y, random_state=1)
```

```python
model= DecisionTreeRegressor(random_state=1)
```

```python
model.fit(train_X,train_y)
```

```python
model.predict(val_X)
```

```python
predicted=model.predict(val_X)
```

```python
mae=val_mae = mean_absolute_error(predicted, val_y)
```
